```python
import os
from rouge_score import rouge_scorer
from collections import defaultdict

# Define the paths to your reference files (eval files)
reference_path = r"C:\Users\ASUS\OneDrive\Máy tính\npp\legaldoc_summarize\eval"

# Define the path to the hypothesis files (files that need evaluation)
hypothesis_path = r"C:\Users\ASUS\OneDrive\Máy tính\npp\legaldoc_summarize\comBI"

# Function to read the content of each file
def read_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()

# Initialize the ROUGE scorer
scorer = rouge_scorer.RougeScorer(["rouge1", "rouge2", "rougeL"])

# Initialize a dictionary to store ROUGE scores for each file
rouge_scores = defaultdict(list)

# List of file names (you can add more or use os.listdir() to get the files dynamically)
file_names = [
    "10186_combined_summarize",
    "10317_combined_summarize",
    "10418_combined_summarize",
    "10461_combined_summarize",
    "10564_combined_summarize",
    "10015_combined_summarize",
    "10047_combined_summarize",
    "10106_combined_summarize",
    "10135_combined_summarize",
    "10185_combined_summarize"
]

# Loop through the files and evaluate ROUGE
for file_name in file_names:
    reference_file = os.path.join(reference_path, f"{file_name}.txt")  # Reference file (eval file)
    hypothesis_file = os.path.join(hypothesis_path, f"{file_name}.txt")  # Hypothesis file (to evaluate)

    # Print out full paths for debugging
    print(f"Checking for reference file: {os.path.abspath(reference_file)}")
    print(f"Checking for hypothesis file: {os.path.abspath(hypothesis_file)}")
    
    if os.path.exists(reference_file) and os.path.exists(hypothesis_file):  # Check if both files exist
        reference_text = read_file(reference_file)  # Read reference file
        hypothesis_text = read_file(hypothesis_file)  # Read hypothesis file
        
        # Calculate ROUGE score
        scores = scorer.score(reference_text, hypothesis_text)
        
        # Store the ROUGE scores
        rouge_scores[file_name].append(scores["rouge1"].fmeasure)
        rouge_scores[file_name].append(scores["rouge2"].fmeasure)
        rouge_scores[file_name].append(scores["rougeL"].fmeasure)
    else:
        print(f"Files for {file_name} not found. Check the file paths.")
        if not os.path.exists(reference_file):
            print(f"Reference file {file_name} is missing.")
        if not os.path.exists(hypothesis_file):
            print(f"Hypothesis file {file_name} is missing.")

# Display the results
for file_name, scores in rouge_scores.items():
    print(f"ROUGE Scores for {file_name}:")
    print(f"  ROUGE-1: {scores[0]:.4f}")
    print(f"  ROUGE-2: {scores[1]:.4f}")
    print(f"  ROUGE-L: {scores[2]:.4f}")
    print()
```
